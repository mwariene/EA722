### Experiência 2 - Emulador Industrial


Para fazer a análise do erro, precisamos analisar a região do gráfico em que a curva deveria se estabilizar então, utilizaremos a região próxima de 1 segundo.

No gráfico do sistema sem perturbação, ao olharmos para essa região, é possível notar que todas as três curvas coloridas acabam convergindo e se sobrepondo exatamente em cima da linha tracejada nos 4000. Isso nos mostra  que o erro de regime é nulo para qualquer valor de ganho. Nesse cenário, alterar o valor do ganho afeta apenas o formato inicial da curva, ou seja, o regime transitório. Podemos ver que o ganho maior, representado pela curva verde, faz o sistema subir mais rápido e ultrapassar mais a referência no início (maior sobressinal), mas no final, todas as configurações atingem o objetivo perfeitamente.

Já no segundo gráfico, onde a perturbação foi introduzida, a leitura visual do erro na mesma região de estabilização antes de 1 segundo, fica claro que nenhuma das curvas consegue alcançar e se manter na linha tracejada dos 4000. Elas estabilizam em níveis diferentes, acima da referência, o que comprova a presença de um erro contínuo causado pela mudança na dinâmica da planta. Ao olharmos para as curvas, identificamos que quanto menor o ganho, mais distante da região esperada, ou seja, maior o erro.
